```python
import warnings

# Filter out the InsecureRequestWarning
warnings.filterwarnings("ignore")

from elasticsearch import Elasticsearch
import pandas as pd
es = Elasticsearch(
    "https://localhost:9200",
    basic_auth=("elastic", "0AvRKA69Ud34+k2Uphbn"),
    verify_certs=False
)
es.ping()
```

```python
df = pd.read_csv("imdb_top_1000.csv")
df.head()
```

```python
!pip install sentence_transformers
from sentence_transformers import SentenceTransformer
model = SentenceTransformer("all-mpnet-base-v2")
```

```python
columns_to_concatenate = ['Series_Title','Genre','Overview','Director','Star1','Star2','Star3','Star4']

# Concatenate columns into a new column named 'concatenated'
df['concatenated'] = df[columns_to_concatenate].astype(str).apply(' '.join, axis=1)
```

```python
df.head()
```

```python
df["vector_embedding"] = df["concatenated"].apply(lambda x: model.encode(x))
```

```python
df.head()
```

```python
indexMapping = {
    "properties":{
        "Series_Title":{
            "type":"text"
        },
        "Genre":{
            "type":"text"
        },
        "Overview":{
            "type":"text"
        },
        "Director":{
            "type":"text"
        },
        "Star1":{
            "type":"text"
        },
        "Star2":{
            "type":"text"
        },
        "Star3":{
            "type":"text"
        },
        "Star4":{
            "type":"text"
        },
        "concatenated":{
          "type":"text"  
        },
        "vector_embedding":{
            "type":"dense_vector",
            "dims": 768,
            "index": True,
            "similarity": "cosine"
        }
        
    }
}
```

```python
es.indices.delete(index="vector_db")
```

```python
es.indices.create(index="vector_db", mappings=indexMapping)
```

```python
import warnings

# Filter out the InsecureRequestWarning
warnings.filterwarnings("ignore")

for idx, row in df.iterrows():
    try:

        document = {
            'Series_Title': row['Series_Title'],
            'Released_Year': row['Released_Year'],
            'Genre': row['Genre'],
            'Overview': row['Overview'],
            'Director': row['Director'],
            'Star1': row['Star1'],
            'Star2': row['Star2'],
            'Star3': row['Star3'],
            'Star4': row['Star4'],
            'concatenated': row['concatenated'],
            'vector_embedding': row['vector_embedding']   # Convert embedding vector to list
        }

        # Index the document into Elasticsearch
        es.index(index="vector_db", document=document, id=idx)
        print(f"Record {idx} indexed successfully!")
    except Exception as e:
        print(f"Error indexing record {idx}:", e)

```
